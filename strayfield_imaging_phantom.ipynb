```python
import nmrglue  # use version 0.5
import pylab
import os
from scipy import fftpack
%matplotlib inline
```

```python
expt_no = "24"

directory = os.path.join(  "eeh_20150527_imaging" )
file_rr = os.path.join( directory, expt_no, "pdata", "1")
file_ser = os.path.join(directory,expt_no)
vd_file  = os.path.join(directory, expt_no, "vdlist")

print directory
```

```python
info,expt_fid = nmrglue.bruker.read( file_ser )
fid_B = abs(expt_fid)
print fid_B.shape
```

```python
pylab.plot(expt_fid.real);
pylab.xlabel("points");
pylab.title("FID showing 25 echoes of 240 pts");
```

```python
pylab.plot(expt_fid.real[0:240*4]);
pylab.xlabel("points");
pylab.title("Expansion of FID showing first 4 echoes");
```

```python
fid1 = expt_fid[0+240*0:0+240*25]  # chop fid down to a certain number of echoes
```

```python
fid1 = fid1.reshape((25,240))   # reshape fid into a matrix of echoes
```

```python
pylab.plot(fid1[3].real);
pylab.plot(1.7*fid1[10].real);
pylab.xlabel("points");
pylab.title("Overlap of two Echoes");
```

```python
# Sum Echoes and Display

sss = fid1.sum(axis=0)
pylab.plot(sss.real);
pylab.plot(sss.imag);

pylab.xlabel("points");
pylab.title("Summed Echoes");
```

```python
iii = fftpack.fft(sss)  # FFT echo, resulting array will be a power of 2 long
```

```python
iii = fftpack.fftshift(iii) # Shift 0 into the middle
```

```python
pylab.plot((iii.real));
pylab.plot(iii.imag+1e8);

pylab.xlabel("points");
pylab.title("Real and Imaginary Data of FFT Image");
```

```python
pylab.plot(abs(iii));
pylab.xlabel("points");
pylab.title("Magnitude Representation of FFT Image");
```

```python
pylab.plot(abs(iii[75:180]));
pylab.xlabel("points");
pylab.title("Exapnsion of Magnitude\n Representation of FFT Image");
```
