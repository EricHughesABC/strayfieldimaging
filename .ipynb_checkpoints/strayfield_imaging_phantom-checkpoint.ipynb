```python
import numpy
import scipy
import nmrglue
import pylab
import os
from scipy import fftpack
%matplotlib inline
```

```python
expt_no = "24"

directory = os.path.join(  "eeh_20150527_imaging" )

file_rr = os.path.join( directory, expt_no, "pdata", "1")

file_ser = os.path.join(directory,expt_no)

vd_file  = os.path.join(directory, expt_no, "vdlist")


print directory
```

```python
info,expt_fid = nmrglue.bruker.read( file_ser )
fid_B = abs(expt_fid)
print fid_B.shape
```

```python
pylab.plot(expt_fid.real);
```

```python
pylab.plot(expt_fid.real[0:240*4]);
```

```python
fid1 = expt_fid[0+240*0:0+240*25]
```

```python
fid1 = fid1.reshape((25,240))
```

```python
pylab.plot(fid1[3].real);
pylab.plot(1.7*fid1[10].real);
```

```python
sss = fid1.sum(axis=0)
pylab.plot(sss.real);
pylab.plot(sss.imag);
```

```python
iii = fftpack.fft(sss)
```

```python
iii = fftpack.fftshift(iii)
```

```python
pylab.plot((iii.real));
pylab.plot(iii.imag+1e8);

```

```python
pylab.plot(abs(iii));
```
